```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
ds=pd.read_csv('RTA Dataset.csv')
```

```python
ds.head()
```

```python
ds.isnull().sum()
```

```python
ds.drop(['Fitness_of_casuality', 'Time', 'Service_year_of_vehicle', 'Work_of_casuality', 'Defect_of_vehicle'], axis=1, inplace=True, errors='ignore')
```

```python
ds['Educational_level'] = ds['Educational_level'].fillna('Unkown')
ds['Type_of_vehicle']=ds['Type_of_vehicle'].fillna('Unknow')
ds['Owner_of_vehicle']=ds['Owner_of_vehicle'].fillna('Unknow')
ds['Area_accident_occured']=ds['Area_accident_occured'].fillna('Other')
ds['Lanes_or_Medians']=ds['Lanes_or_Medians'].fillna('Other')
ds['Types_of_Junction']=ds['Types_of_Junction'].fillna('No juncation')
ds['Vehicle_movement']=ds['Vehicle_movement'].fillna('Unknow')
ds['Type_of_collision']=ds['Type_of_collision'].fillna('Unknow')
ds['Road_surface_type']=ds['Road_surface_type'].fillna('Other')
 
```

```python
ds['Vehicle_driver_relation']=ds['Vehicle_driver_relation'].fillna('Unknow')
```

```python
ds['Driving_experience']=ds['Driving_experience'].fillna('Unknow')
ds['Road_allignment']=ds['Road_allignment'].fillna('Unknow')
```
